```python
train_ids = fn_train.id
fn_train = fn_train.drop(['id'], axis =1)
test_ids = fn_test.id
fn_test = fn_test.drop(['id'], axis =1)
```

```python
x = fn_train.drop('species',axis=1)
y = fn_train['species']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_fit = le.fit(fn_train['species'])
y_label = y_fit.transform(fn_train['species']) 
classes = list(y_fit.classes_) 
```

```python
from sklearn.model_selection import  train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y_label, test_size = 0.2, random_state =1)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 40)
classifier.fit(x_train, y_train)
```

```python
from sklearn.metrics import classification_report
predictions = classifier.predict(x_test)
print (classification_report(y_test, predictions))
```

```python
final_predictions = classifier.predict_proba(fn_test)
submission = pd.DataFrame(final_predictions, columns=classes)
submission.insert(0, 'id', test_ids)
submission.reset_index()
```
